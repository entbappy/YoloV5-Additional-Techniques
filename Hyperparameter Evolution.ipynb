## Define Fitness

``def fitness(x):
    # Model fitness as a weighted combination of metrics
    w = [1.0, 1.0, 0.2, 0.8]  # weights for [P, R, mAP@0.5, mAP@0.5:0.95]
    return (x[:, :4] * w).sum(1)``


You can assign how much weight you want for each parameters.
